**IMPORTING OUR MODULE**

```python
from Qvntvs import Qvntvs

particle = Qvntvs(potential_barrier_electron=1000,
                            potential_barrier_hole=1000,
                            m_e_barrier=1, 
                            m_e_well = 1,
                            m_h_barrier=1, 
                            m_h_well=1, 
                            biasing_voltage=3,
                            well_width_nm=5, 
                            barrier_width_nm=10, 
                            n_intervals=1000)
```

**DEFINING THE POTENTIALS FOR ELECTRONS AND HOLES**

```python
V_general_electron = particle.rectangular_potential_profile(electron=True, plot=True)
V_general_hole = particle.rectangular_potential_profile(electron=False, plot=True)
```

**DEFINING THE EFFECTIVE AND INVERSE MASSES**

```python
m_general_hole = particle.effective_mass_profile(electron=False, plot=False)
m_general_electron = particle.effective_mass_profile(electron=True, plot=False)

inv_mass_hole = particle.inverse_mass_profile(m_general_hole, electron=False, plot=False)
inv_mass_electron = particle.inverse_mass_profile(m_general_electron, electron=True, plot=False)
```

**DEFINING HAMILTONIAN MATRIX AND PLOTTING THE SOLUTIONS**

```python
H_electron = particle.hamiltonian_matrix(V_general_electron, inv_mass=inv_mass_electron, electron=True, plot=True)
H_hole = particle.hamiltonian_matrix(V_general_hole, inv_mass=inv_mass_hole, electron=False, plot=True)

energy_levels_electron, wave_functions_electron = particle.eigen_equation(H_electron, electron=True, plot=True, n_levels=4)
energy_levels_hole, wave_functions_hole = particle.eigen_equation(H_hole, electron=False, plot=True, n_levels=4)
```
